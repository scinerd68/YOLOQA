```python
import pickle

with open('data/train_preprocessed2.pkl', 'rb') as f:
    train = pickle.load(f)
```

```python
with open('data/debug_preprocessed2.pkl', 'rb') as f:
    debug = pickle.load(f)
```

```python
debug.keys()
```

```python
train.keys()
```

```python
train['input_ids'][0][0]
```

```python
for example_idx, start in enumerate(train['start_positions']):
    if start == 512:
        print(example_idx)
        edge_case = {key: val[example_idx] for key, val in train.items()}
        print(len(edge_case['input_ids']))
        break
```

```python
edge_case['end_positions']
```

```python
for key in debug:
    debug[key].insert(0, edge_case[key])
```
